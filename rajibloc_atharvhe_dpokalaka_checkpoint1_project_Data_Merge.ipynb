```python
import pandas as pd

# Load the dataset
df = pd.read_csv('aime.csv')
```

```python
df
```

```python
df = df.loc[:, ["questions", "answers"]]
```

```python
df = df.rename(columns={'answers': 'solutions'}) 
```

```python
df['solutions'] = df['solutions'].astype(str)
```

```python
import re

def extract_boxed_value(solution):
    match = re.search(r'\\boxed{(\d+)}', solution)
    if match:
        return match.group(1)
    return None
```

```python
df['answers'] = df['solutions'].apply(extract_boxed_value)
```
